### MATPLOTLIB

```python

# 1
furnishing_counts = house_data['furnishing'].value_counts()
plt.pie(furnishing_counts, labels=furnishing_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Furnishing Types')
plt.axis('equal')
plt.show()

```

```python
# 2
top_locations = house_data.groupby('location')['price__in_rupees_'].mean().sort_values(ascending=False).head(10)
plt.bar(top_locations.index, top_locations.values, color='green')
plt.xticks(rotation=45)
plt.title('Top 10 Locations by Avg Price')
plt.xlabel('Location')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()

```

```python
# 3
plt.scatter(house_data['carpet_area'], house_data['price__in_rupees_'], alpha=0.5)
plt.title('Price vs Carpet Area')
plt.xlabel('Carpet Area (sqft)')
plt.ylabel('Price (in rupees)')
plt.show()
```

```python
# 4
ownership_counts = house_data['ownership'].value_counts()
ownership_counts.plot(kind='bar', color='purple')
plt.title('Ownership Type Count')
plt.xlabel('Ownership')
plt.ylabel('Number of Properties')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pdfplumber
from transformers import pipeline

# Step 1: Extract text from PDF
pdf_path = "SIH-2025_Problem_Statements.pdf"
text_data = ["AI"]

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text_data.append(page.extract_text())

all_text = "\n".join(text_data)

# Step 2: Filter topics you want
# Example: Select only "AI", "Blockchain", "Smart Education"
keywords = ["AI", "Blockchain", "Smart Education"]
selected_topics = []

for line in all_text.split("\n"):
    if any(kw.lower() in line.lower() for kw in keywords):
        selected_topics.append(line)

# Step 3: Summarize
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

summaries = []
for topic in selected_topics:
    summary = summarizer(topic, max_length=40, min_length=15, do_sample=False)
    summaries.append(summary[0]['summary_text'])

# Output
for i, (topic, summary) in enumerate(zip(selected_topics, summaries), 1):
    print(f"{i}. Original: {topic}\n   Summary: {summary}\n")
```
